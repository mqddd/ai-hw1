```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.decomposition import PCA
```

```python
df = pd.read_csv('../data/Mall_Customers.csv')
df.head()
```

```python
df.drop(columns='CustomerID', inplace=True)
df.head()
```

```python
df.describe()
```

```python
x_train_without_genre = df.values[:, 1:4]

scaler = StandardScaler(with_mean=True, with_std=True)
scaler.fit(x_train_without_genre)
normalized_x_train_without_genre = scaler.transform(x_train_without_genre)

le = LabelEncoder()
df['Genre'] = le.fit_transform(df['Genre'])

x_train = np.hstack((df['Genre'].to_numpy().reshape(200, 1), normalized_x_train_without_genre))
x_train[0:5]
```

```python
def train_kmeans_model(x, k):
    kmeans = KMeans(n_clusters=k, random_state=None)
    kmeans.fit(x)
    return kmeans
```

```python
inertias = []
for k in range(2, 11):
    kmeans = train_kmeans_model(x_train, k)
    inertias.append(kmeans.inertia_)
```

```python
plt.plot(range(2, 11), inertias, marker = 'o')
plt.show()
```

```python
kmeans_for_best_k = train_kmeans_model(x_train, k=4)

test_df = pd.read_csv('../data/Mall_test.csv')
test_df.head()
```

```python
test_df.drop(columns='Unnamed: 0', inplace=True)
x_test_without_genre = test_df.values[:, 1:4]

normalized_x_test_data_without_genre = scaler.transform(x_test_without_genre)

test_df['Genre'] = le.fit_transform(test_df['Genre'])
x_test = np.hstack((test_df['Genre'].to_numpy().reshape(50, 1), normalized_x_test_data_without_genre))
x_test[0:5]
```

```python
predicted_test_data_cluster_without_dimension_reduction = kmeans_for_best_k.predict(x_test)
df_for_test_data = pd.DataFrame(x_test)
df_for_test_data['predict1'] = predicted_test_data_cluster_without_dimension_reduction
df_for_test_data.to_csv('predicts.csv')
df_for_test_data.head()
```

```python
pca = PCA(n_components=2)
pca.fit(x_train)
x_train_proj = pca.transform(x_train)
x_test_proj = pca.transform(x_test)
pca_train_1 = x_train_proj[:, 0]
pca_train_2 = x_train_proj[:, 1]
pca_test_1 = x_test_proj[:, 0]
pca_test_2 = x_test_proj[:, 1]

kmeans_with_dimension_reduction = train_kmeans_model(x_train_proj, k=4)
predicted_train_data_cluster_with_dimension_reduction = kmeans_with_dimension_reduction.predict(x_train_proj)
predicted_test_data_cluster_with_dimension_reduction = kmeans_with_dimension_reduction.predict(x_test_proj)
df_for_test_data['predict2'] = predicted_test_data_cluster_with_dimension_reduction
df_for_test_data.to_csv('predicts.csv')
df_for_test_data.head()
```

```python
plt.figure(figsize=(5, 5))

plt.scatter(pca_train_1, pca_train_2, c=predicted_train_data_cluster_with_dimension_reduction, cmap='plasma', marker='o', alpha=0.5)
plt.scatter(pca_test_1, pca_test_2, c=predicted_test_data_cluster_with_dimension_reduction, cmap='plasma', marker='<')
plt.xlabel('first component')
plt.ylabel('second component')
centers = kmeans_with_dimension_reduction.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, marker='+')

plt.show()
```
